** barplot and countplot**

```python
# to find mean value of total_bill for smoker and nonsmoker
sns.catplot(x='smoker', y='total_bill', kind='bar', data=df)
```

```python
# adding a hue parameter
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=df)
```

```python
# bydefualt estimator value is mean but we can change it
# np.mean , np.std, np.median, np. var
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.std, data=df)
```

```python
# count plots :
# kind of bar plots which wokrs on categorical data, count of that category  
```

```python
sns.catplot(x='sex', kind='count', data=df)
```

```python
# adding the hue parameter
sns.catplot(x='sex', kind='count', hue='smoker',data=df)
```

```python
# heat maps
# graphical repersentation where individual values -
# contained in a matrix are reperesented as colour

# importing new datset 
df2=sns.load_dataset('flights')
df2.head()
```

```python
# coverting it in to pivot table 
x= df2.pivot_table(index='year', columns='month', 
                values='passengers', aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
# adding line with btwn the boxes and 
# adding numerical annotations of 'd' integer values
# changing the colour tone viridis, summer, winter
sns.heatmap(x, linewidth=0.5, annot=True,fmt='d', cmap='viridis')
```

#distpolts
distplots are used to do univarient analysis on continues/numerical data
1. histogram
2. kernal density function(pdf)
3. Rugplot


```python
df5=sns.load_dataset('titanic')
df5
```

```python
df5['age'].fillna(df5['age'].mean(), inplace= True)
# filling empty values in age with mean value
```

```python
# plotting distplot 
# histogram +kde
sns.displot(df5['age'], kde= True)
```

```python
#plotting only kde
sns.displot(df5[df5['survived']==1]['age'], kde= True) 
sns.displot(df5[df5['survived']==0]['age'], kde= True)

```
